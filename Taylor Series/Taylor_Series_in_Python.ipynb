# Use a *for loop* to calculate a Taylor Serie

Si queremos acercarnos al valor de $e^{x}$, necesitamos agregar más términos a nuestra serie de Taylor. El problema es que codificar cada término individual requiere mucho tiempo y es repetitivo. En lugar de codificar cada término individualmente, podemos usar un bucle for. Un bucle for es una estructura de repetición en Python que ejecuta una sección de código un número específico de veces. La sintaxis para codificar un bucle for en Python usando la función $range ()$ es la siguiente:

for $<var>$ in $range(<num>)$
  $<code>$

Donde $<var>$ es cualquier nombre de variable de Python válido y $<num>$ es un número entero para determinar cuántas veces se ejecuta el $<code>$ en el bucle.

Podemos recrear nuestra aproximación de $e^{2}$ con 5 términos usando un ciclo for. Tenga en cuenta que debemos establecer la variable **e_to_2** en $0$ antes de que comience el ciclo. El operador matemático $+ =$ en la línea **e_to_2** $+ = x ** i / math.factorial (i)$ es equivalente a **e_to_2**$ = e_to_2 + x ** i / math.factorial (i).$


```
import math

x = 2
e_to_2 = 0
for i in range(5):
    e_to_2 += x**i/math.factorial(i)
    
print(e_to_2)
```

```
import math

x = 2
sin_to_2 = 0
for i in range(10):
    sin_to_2 +=(-1)**i * x**(2*i+1)/math.factorial(2*i+1) 
print(sin_to_2)
```

```
x = 2
cos_to_2 = 0
for i in range(10):
    cos_to_2 +=(-1)**i * x**(2*i)/math.factorial(2*i) 
print(cos_to_2)
```

```
x = 0.5
frac_5 = 0
for i in range(10):
    frac_5 +=x**i
print(frac_5)
```

```
x = 0.5
ln_1_x = 0
for i in range(10):
    ln_1_x +=(-1)**(i)*x**(i+1)/(i+1) 
print(ln_1_x)
```

```
x = 0.5
ln_x = 0
for i in range(10):
    ln_x +=(-1)**(i)*(x-1)**(i+1)/(i+1) 
print(ln_x)
```

El resultado 7.0 es el mismo que el resultado que calculamos cuando escribimos cada término de la serie de Taylor individualmente.

Una ventaja de usar un bucle for es que podemos aumentar fácilmente el número de términos. Si aumentamos el número de veces que se ejecuta el bucle for, aumentamos el número de términos en la expansión de la serie Taylor. Probemos con $10$ términos. Observe cómo la línea **for i in range(10):** ahora incluye $10$ pasado a la función $range ().$

```
import math

x = 2
e_to_2 = 0
for i in range(10):
    e_to_2 += x**i/math.factorial(i)
    
print(e_to_2)
```

# Create a function to estimate the value of cos(x) using a Taylor Series

Next, let's calculate the value of the cosine function using a Taylor Series. The Taylor Series expansion for **cos(x)** is below.

Podemos codificar esta fórmula en una función que contenga un ciclo for. Tenga en cuenta que la variable x es el valor del que estamos tratando de encontrar el coseno, la variable n es el número de términos en la serie de Taylor y la variable i es el índice de bucle, que también es el número de término de la serie de Taylor. Estamos usando una variable separada para el coeficiente **coef** que es igual a $(-1)^{i}$, el numerador **num** que es igual a $x^{2i}$ y el denominador **denom** que es igual a **(2i!)**. Dividir la fórmula de la serie Taylor en tres partes puede reducir los errores de codificación.

```
import math

def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )
    
    return cos_approx
```

```
import math

def func_sin(x, n):
    sin_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i +1 )
        denom = math.factorial(2*i + 1)
        sin_approx += ( coef ) * ( (num)/(denom) )
    
    return sin_approx
```

```
import math

def func_e(x, n):
    e_approx = 0
    for i in range(n):
        coef = 1
        num = x**(i)
        denom = math.factorial(i)
        e_approx += ( coef ) * ( (num)/(denom) )
    
    return e_approx
```

```
import math

def func_frac(x, n):
    frac_approx = 0
    for i in range(n):
        coef = 1
        num = x**(i)
        denom = 1
        frac_approx += ( coef ) * ( (num)/(denom) )
    
    return frac_approx
```

```
import math

def func_ln1(x, n):
    ln1_approx = 0
    for i in range(n):
        coef = (-1)**(i)
        num = x**(i + 1)
        denom = (i+1)
        ln1_approx += ( coef ) * ( (num)/(denom) )
    
    return ln1_approx
```

```
import math

def func_ln(x, n):
    ln_approx = 0
    for i in range(n):
        coef = (-1)**(i)
        num = (x-1)**(i + 1)
        denom = (i+1)
        ln_approx += ( coef ) * ( (num)/(denom) )
    
    return ln1_approx
```

# Build a plot to compare the Taylor Series approximation to Python's cos() function

En la última parte de esta publicación, vamos a construir un gráfico que muestra cómo la aproximación de la serie Taylor calculada por nuestra función **func_cos ()** se compara con la función **cos ()** de Python.

La idea es hacer una gráfica que tenga una línea para la función **cos ()** de Python y líneas para la aproximación de la serie de Taylor basada en diferentes números de términos.

Por ejemplo, si usamos 3 términos en la aproximación de la serie de Taylor, nuestra gráfica tiene dos líneas. Una línea para la función **cos ()** de Python y una línea para nuestra función **func_cos ()** con tres términos en la aproximación de la serie de Taylor. Calcularemos el coseno usando ambas funciones para ángulos entre −2π radianes y 2π radianes.

```
import math
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_cos = np.cos(angles)
t_cos = [func_cos(angle,3) for angle in angles]

fig, ax = plt.subplots()
ax.plot(angles,p_cos)
ax.plot(angles,t_cos)
ax.set_ylim([-5,5])
ax.legend(['cos() function','Taylor Series - 3 terms'])

plt.show()
```
